```python
import numpy as np 
from pylab import *

np.random.seed(2)

pageSpeeds = np.random.normal(3, 1, 100)
purchaseAmount = np.random.normal(50, 30, 100) / pageSpeeds

scatter(pageSpeeds, purchaseAmount)
```

```python
testX = pageSpeeds[80:]
testY = purchaseAmount[80:]

trainX = pageSpeeds[:80]
trainY = purchaseAmount[:80]

scatter(trainX, trainY)
scatter(testX,testY)
```

```python
x = np.array(trainX)
y = np.array(trainY)

p = np.poly1d(np.polyfit(x,y,8))
```

for train data

```python
import matplotlib.pyplot as plt

xp = np.linspace(0, 7, 100)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])

plt.scatter(x,y)
plt.plot(xp, p(xp), c='r')
plt.show()
```

for test data

```python
testx = np.array(testX)
testy = np.array(testY)

axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])

plt.scatter(testx, testy)
plt.plot(xp, p(xp), c='r')
plt.show()
```

finding R-square value

```python
from sklearn.metrics import r2_score

r2 = r2_score(testy, p(testx))

print(r2)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(np.array(trainY), p(np.array(trainX)))

print(r2)
```
